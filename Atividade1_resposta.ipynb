#ITEM 1: Criar dataset de quadrado, circulo, triangulo.

```
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import random
amostras = 2400

altura = 100
largura = 100
tamanho = altura * largura

f = open("dataset_exe1.csv", "w")

# Cabecalho do dataset.
data = ""
for py in range(0,altura):
  for px in range(0, largura):
    data = data + "P"+str(py)+str(px)+","
data = data + "Saida\n"
f.write( data )

#maximo de lados 
lados_max = 4
qtde_n_lados = [0]*(lados_max+1)
for index in range (0, lados_max):
  qtde_n_lados[index] = 0

n_lados = 2

for amostra in range(0,amostras):
  img = np.zeros([altura,largura,3])
  for py in range(0,altura):
    for px in range(0, largura):
      img[px][py][0] = 255   # cor azul
      img[px][py][1] = 255   # cor verde
      img[px][py][2] = 255   # cor vermelho

  lado = random.randint(50, altura)
  raio = int(lado/2)
  x_min = random.randint(0, largura - lado)
  xc = x_min + raio
  y_min = random.randint(0, altura - lado) 
  yc = y_min + raio

  x0 = x_min
  y0 = yc + int( (raio**2 - (x_min - xc)**2)**0.5)
  circulo_pts = np.array([[x0,y0]])

  for index in range (1, lado):
    x = x_min + index
    y = int( (raio**2 - (x - xc )**2)**0.5) + yc
    pto = np.array([[x,y]])
    circulo_pts = np.concatenate((circulo_pts, pto),axis=0)

  for index in range (lado-1, 1,-1):
    x = x_min + index
    y = (int( -(raio**2 - (x - xc )**2)**0.5) + yc)
    pto = np.array([[x,y]])
    circulo_pts = np.concatenate((circulo_pts, pto),axis=0)

  circulo_pts = circulo_pts.reshape((-1,1,2))
  qtde_n_lados[n_lados] = qtde_n_lados[n_lados] + 1

  if (n_lados == 4):
    cv2.rectangle(img,(x_min, y_min),(x_min+lado,y_min+lado),(0,0,0),-1)
  elif (n_lados == 2):
    x_centro = x_min + raio
    y_centro = y_min + raio
    cv2.circle(img, (xc, yc), raio ,(0,0,0), -1)
  else:
    poligono_indice = [0]*n_lados
    ini = 0
    dx = int ((len(circulo_pts) -1)/n_lados)
    max = dx
    np.sort(poligono_indice)
    for i in range (0,n_lados):
      index = ini
      if ( i== 0 ):
        poligono_pts = np.array([circulo_pts[index]])
      else:
        pto = np.array([circulo_pts[index]])
        poligono_pts = np.concatenate((poligono_pts, pto),axis=0)
      if ( ini < len(circulo_pts) ):
        ini = ini + dx
      else:
        ini = len(circulo_pts)-1
    np.sort(poligono_pts)
    poligono_pts = poligono_pts.reshape((-1,1,2))
    cv2.fillPoly(img, [poligono_pts],(0,0,0))
  
  data = ""
  for py in range(0,altura):
    for px in range(0, largura):
      data = data + str(int(img[px][py][0]))+","
  if ( n_lados == 4):
    data = data + "4\n"
  elif ( n_lados== 2):
    data = data + "0\n"
  else:
    data = data + str(n_lados) +"\n"
  
  print("Amostra: ", amostra," Lados: ",n_lados,end="\n" )
  cv2_imshow(img)
  
  f.write( data )
  if (n_lados == lados_max ):
    n_lados =2
  else:
    n_lados = n_lados + 1

f.close()
for index in range (2, lados_max+1):
  print("QTDE [",index,"]: ",qtde_n_lados[index],end="\n")
```

#ITEM 2: Criar uma rede neural que classifique quadrado, circulo, triangulo, a partir do dataset do item 1.

#LER DATASET

```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# LER DATASET

dados = pd.read_csv("dataset_exe1.csv")
dados.head()
dados_entrada = dados.drop("Saida", axis=1)
dados_saida = dados["Saida"]

# Separacao de dados de treino e teste.
dados_entrada_treino, dados_entrada_teste, dados_saida_treino, dados_saida_teste = train_test_split(dados_entrada, dados_saida,test_size=0.3, random_state = 42)

#REDE NEURAL: TREINO/TESTE
qualidade = 0
#Achar uma qualidade de rede maior que 0.85
while(qualidade < 0.85 ):
  #Treino
  #clf = MLPClassifier( activation='relu',solver='lbfgs', max_iter = 1000)
  clf = MLPClassifier( activation='relu', max_iter = 2000)
  clf = clf.fit( dados_entrada_treino, dados_saida_treino)
  #Teste
  calculado = clf.predict(dados_entrada_teste)
  real = dados_saida_teste.values
  print("Calculado: ",calculado[0:15])
  print("Saida:     ",real[0:15])
  #Qualidade
  qualidade = accuracy_score(real, calculado)
  print (qualidade)
```
